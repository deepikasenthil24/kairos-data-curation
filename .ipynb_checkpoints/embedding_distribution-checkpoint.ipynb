## Dataset Embedding Distribution Comparison

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
# load embedding files
train_embeds = np.load('inat_embeddings.npy') 
val_embeds = np.load('val_embeddings.npy')

# load label files
def load_txt_labels(filepath):
    with open(filepath, 'r') as f:
        return [line.strip() for line in f.readlines()]

train_labels = load_txt_labels('inat_labels.txt')
val_labels = load_txt_labels('val_labels.txt')
```

```python
# combine embeddings
all_embeddings = np.concatenate([train_embeds, val_embeds], axis=0)

# create source labels
sources = (['iNaturalist'] * len(train_embeds)) + (['Validation'] * len(val_embeds))
```

```python
# reduce to 2D
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
```

```python
# fits t-SNE model to data and then returns the data transformed into 2d embedded space
embeddings_2d = tsne.fit_transform(all_embeddings)
```

### Embedding Space Comparison by Dataset

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x=embeddings_2d[:, 0], 
    y=embeddings_2d[:, 1], 
    hue=sources, 
    alpha=0.5,
    palette={'iNaturalist': 'blue', 'Validation': 'orange'}
)
plt.title("Embedding Space: iNaturalist vs. Validation Set")
plt.xlabel("t-SNE Dimension 1", fontsize=10)
plt.ylabel("t-SNE Dimension 2", fontsize=10)
plt.show()
```

We can see some overlap between the two datasets, confirming that KAIROS has a bridge to work with to find high-value points.

### Embedding Space Comparison by Class

```python
unique_labels = sorted(list(set(train_labels + val_labels)))

# define color palette
colors = sns.color_palette("tab10", len(unique_labels))
color_palette = {}

for i, label in enumerate(unique_labels):
    if label.lower() == 'noise':
        color_palette[label] = "#D3D3D3" # light grey
    else:
        color_palette[label] = colors[i]
```

```python
train_2d = embeddings_2d[:len(train_embeds)]
val_2d = embeddings_2d[len(train_embeds):]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8), sharex=True, sharey=True)

# plot iNaturalist
sns.scatterplot(x=train_2d[:,0], y=train_2d[:,1], hue=train_labels, 
                ax=ax1, palette=color_palette, alpha=0.9, s=20)
ax1.set_title("iNaturalist Embedding Distribution by Class", fontsize=15, pad=10)
ax1.set_xlabel("t-SNE Dimension 1", fontsize=12)
ax1.set_ylabel("t-SNE Dimension 2", fontsize=12)

# to remove redundant legend from first subplot
if ax1.get_legend():
    ax1.get_legend().remove()

# plot validation
sns.scatterplot(x=val_2d[:,0], y=val_2d[:,1], hue=val_labels, 
                ax=ax2, palette=color_palette, alpha=0.9, s=60)
ax2.set_title("Validation Embedding Distribution by Class", fontsize=15, pad=10)
ax2.set_xlabel("t-SNE Dimension 1", fontsize=12)

# create a single, clean legend for the whole figure
handles, labels = ax1.get_legend_handles_labels()
ax2.get_legend().remove()

fig.legend(handles, labels, loc='center right', title="Species", 
            title_fontsize='15', fontsize='14', frameon=True)

legend = fig.legend(handles, labels, loc='center right', title="Species", 
                    title_fontsize='15', fontsize='14', frameon=True)

for handle in legend.legend_handles:
    handle.set_alpha(1.0)
    handle.set_markersize(14)

# 3. Prevent the legend from being cut off
plt.subplots_adjust(right=0.85)

plt.tight_layout()
plt.subplots_adjust(right=0.88) 

plt.show()
```

Review these points with Matilda

* The large grey sea of noise in the iNaturalist plot proves that the iNaturalist dataset is highly diluted with non-insect images.
* We see some insect points in the iNaturalist plot directly inside the cloud of grey noise. KAIROS should assign these specific points a low score to prevent the ResNet-50 from learning incorrect patterns (confusing background noise for an insect).
* If a colorful cluster is in the same region in both plots, the visual characteristics (lighting, camera angles) are similar.
* A successful KAIROS run will select the colorful dots that physically overlap with the validation clusters.
